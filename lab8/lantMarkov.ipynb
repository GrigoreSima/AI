```python
from random import randint, random
class Markov():
    _states = 1
    _file = "data/corpus_complet.txt"
    _chain = {}

    def __init__(self, states = 1):
        self._states = states
        self._learn()

    def _learn(self):
        with open(self._file) as file:

            lastKey = ""
            line = file.readline()
            
            while line != "":
                words = line.split(" ")
                while len(words) < self._states:
                    line += " "
                    line += file.readline()
                    words = line.split(" ")


                words = line.split(" ")
                for i in range(len(words)):
                    words[i] = words[i].strip()
                    words[i] = words[i].lower()
                    # words[i] = words[i].replace("...", "")

                if lastKey != "":
                    
                    for i in range(self._states):
                        if i >= len(words):
                            break

                        word = words[i]
                        if lastKey in self._chain:
                
                            forKey = self._chain[lastKey]
                            if word in forKey:
                                forKey[word] += 1
                            else:
                                forKey[word] = 1
                
                        else: 
                            self._chain[lastKey] = {word : 1}
                  
                        index = lastKey.find(" ")
                        if index != -1:
                            lastKey = lastKey[index+1:] + " " + word
                        else:
                            lastKey = word

                for i in range(self._states, len(words)):

                    word = words[i]
                    key = " ".join(words[i - self._states:i])

                    if key in self._chain:
                        forKey = self._chain[key]
                        if word in forKey:
                            forKey[word] += 1
                        else:
                            forKey[word] = 1
                    else: 
                        self._chain[key] = {word : 1}

                lastKey = " ".join(words[len(words) - self._states:])
                line = file.readline()

        for key in self._chain:
            key = self._chain[key]
            total = sum(key.values())

            lastEntry = None
            for entry in key:
                if lastEntry != None:
                    key[entry] = round(key[entry] / total, 2)
                    key[entry] += lastEntry
                else:
                    key[entry] = round(key[entry] / total, 2)

                lastEntry = key[entry]

    
    def generate(self, length = 10):
        result = list(self._chain.keys())[randint(0, len(self._chain) - 1)]
        
        lastWord = result
        for _ in range(length - self._states):

            if lastWord not in self._chain:
               lastWord = list(self._chain.keys())[randint(0, len(self._chain) - 1)]
            
            rand = random()
            word = None

            for key in self._chain[lastWord]:
                word = key
                if self._chain[lastWord][key] > rand:
                    break

            index = lastWord.find(" ")
            if index != -1:
                lastWord = lastWord[index+1:] + " " + word
            else:
                lastWord = word

            result += " "
            result += word

        print(result)
        # print(len(result.split(" ")))

print("Markov with 1 states:")
model = Markov()
model.generate(20)

print()
print()

print("Markov with 3 states:")

model = Markov(3)
model.generate(20)
```
